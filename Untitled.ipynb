```python
"""Q15. Prime Number Check
Take a number from the user and check whether it is prime using for-else."""
# Take input from the user
num = int(input("Enter a number: "))

# Numbers less than 2 are not prime
if num < 2:
    print(f"{num} is not a prime number.")
else:
    # Check for factors from 2 up to the square root of num
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            print(f"{num} is not a prime number.")
            break
    else:
        # This executes if the loop finishes without breaking
        print(f"{num} is a prime number.")
```

```python
"""Q20. Pyramid Pattern
Print:
 *
 ***
 *****
*******
*********"""
# Loop for 5 rows
for i in range(1, 6):
    # Print spaces (decreasing as rows increase)
    print(" " * (5 - i), end="")
    # Print stars (odd numbers: 1, 3, 5, 7, 9)
    print("*" * (2 * i - 1))
```

```python
"""Q21. Inverted Pyramid
Print:
*********
*******
 *****
 ***
 *"""
# Outer loop runs from 5 down to 1
for i in range(5, 0, -1):
    # Print leading spaces (increases as rows decrease)
    print(" " * (5 - i), end="")
    # Print stars (odd numbers: 9, 7, 5, 3, 1)
    print("*" * (2 * i - 1))
```

```python
"""Bonus Question
Q22. Break in Pattern
Print a star pattern.
Stop printing when the row number reaches 4."""
# Loop to print the star pattern
for i in range(1, 6):
    # Check if the row number has reached 4
    if i == 4:
        break
    
    # Print the stars for the current row
    print("*" * i)
```

```python
# 8. Count number of digits in a given number.
num = int(input("Enter a number: "))
count = 0
# Handle the case where the user enters 0
if num == 0:
    count = 1
else:
    # Make the number positive if it's negative
    n = abs(num)
    while n > 0:
        count += 1
        n = n // 10
print(f"The number of digits in {num} is {count}")
```

```python
# 9. Reverse a number using while loop.
num = int(input("Enter a number: "))
reversed_num = 0
# Store the original sign of the number
sign = -1 if num < 0 else 1
n = abs(num)
while n > 0:
    remainder = n % 10
    reversed_num = (reversed_num * 10) + remainder
    n = n // 10
# Restore the original sign
reversed_num *= sign
print(f"The reversed number is: {reversed_num}")
```

```python
# 10. Check whether a number is palindrome or not using while loop.
num = int(input("Enter a number: "))
original_num = num
reversed_num = 0
n = abs(num)
while n > 0:
    remainder = n % 10
    reversed_num = (reversed_num * 10) + remainder
    n = n // 10
# Adjust for negative numbers if necessary
if num < 0:
    reversed_num = -reversed_num
if original_num == reversed_num:
    print(f"{original_num} is a palindrome.")
else:
    print(f"{original_num} is not a palindrome.")
```

```python
"""14. Create a number guessing game:
• Generate a random number (1–10)
• Keep asking user until they guess correctly"""
import random
# Generate a random number between 1 and 10
secret_number = random.randint(1, 10)
guess = 0
print("Welcome to the Number Guessing Game!")
print("I'm thinking of a number between 1 and 10.")
# Keep asking user until they guess correctly
while guess != secret_number:
    guess = int(input("Enter your guess: "))
    
    if guess < secret_number:
        print("Too low! Try again.")
    elif guess > secret_number:
        print("Too high! Try again.")
    else:
        print("Congratulations! You guessed the right number.")
```

```python
#16. Print Fibonacci series up to N terms using while loop.
n_terms = int(input("Enter the number of terms: "))

# First two terms
n1 = 0
n2 = 1
count = 0

# Check if the number of terms is valid
if n_terms <= 0:
    print("Please enter a positive integer.")
elif n_terms == 1:
    print(f"Fibonacci sequence up to {n_terms} term:")
    print(n1)
else:
    print("Fibonacci sequence:")
    while count < n_terms:
        print(n1, end=" ")
        
        # Calculate the next term
        nth = n1 + n2
        
        # Update values for the next iteration
        n1 = n2
        n2 = nth
        
        count += 1
```

```python
#17. Check whether a number is Armstrong number.
num = int(input("Enter a number: "))

# Find the number of digits
temp = num
num_digits = 0
while temp > 0:
    num_digits += 1
    temp //= 10

# Calculate the sum of power of digits
sum_of_powers = 0
temp = num
while temp > 0:
    digit = temp % 10
    sum_of_powers += digit ** num_digits
    temp //= 10

# Check if the number is an Armstrong number
if num == sum_of_powers:
    print(f"{num} is an Armstrong number.")
else:
    print(f"{num} is not an Armstrong number.")
```

```python
#18. Print prime numbers between 1 to 50 using while loop.
i = 2
print("Prime numbers between 1 and 50 are:")

while i <= 50:
    is_prime = True
    j = 2
    
    # Check if i is divisible by any number from 2 up to i // 2
    while j <= i // 2:
        if i % j == 0:
            is_prime = False
            break  # Not a prime number, exit the inner loop
        j += 1
        
    if is_prime:
        print(i, end=" ")
        
    i += 1
print()  # For a new line at the end
```
