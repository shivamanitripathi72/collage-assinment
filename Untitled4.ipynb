```python
# 47. Validate a date using conditions.
day = 29
month = 2
year = 2026 # 2026 leap year nahi hai, isliye 29 Feb invalid hoga
# 2. Leap Year Check (using and, or, not)
# Leap year tab hota hai jab year 4 se divide ho AND (100 se divide NOT ho OR 400 se divide ho)
is_leap = (year % 4 == 0) and (not (year % 100 == 0) or (year % 400 == 0))
# 3. Months ki validity check karna (using or)
is_31_day_month = (month == 1) or (month == 3) or (month == 5) or (month == 7) or (month == 8) or (month == 10) or (month == 12)
is_30_day_month = (month == 4) or (month == 6) or (month == 9) or (month == 11)
is_february = (month == 2)
# 4. Day Validation Conditions (using and, or, not)
# Condition A: 31 din wale mahine me day 1 se 31 ke beech hona chahiye
valid_31 = is_31_day_month and (day >= 1 and day <= 31)
# Condition B: 30 din wale mahine me day 1 se 30 ke beech hona chahiye
valid_30 = is_30_day_month and (day >= 1 and day <= 30)
# Condition C: February me agar leap year hai toh 1-29, agar NOT leap year hai toh 1-28
valid_feb = is_february and ((is_leap and (day >= 1 and day <= 29)) or (not is_leap and (day >= 1 and day <= 28)))
# 5. Final Check (using and, or)
# Date tabhi valid hogi jab Year > 0 ho AND (valid_31 Sach ho OR valid_30 Sach ho OR valid_feb Sach ho)
if (year > 0) and (valid_31 or valid_30 or valid_feb):
    print(f"Result: {day}/{month}/{year} is a VALID date.")
else:
    print(f"Result: {day}/{month}/{year} is an INVALID date.")
```

```python
# 51. Write a Python program to calculate income tax using slabs.
income = 85000
tax = 0
if income <= 10000:
    tax = 0
elif income <= 50000:
    tax = (income - 10000) * 0.10
elif income <= 100000:
    tax = (40000 * 0.10) + (income - 50000) * 0.20
else:
    tax = (40000 * 0.1) + (50000 * 0.2) + (income - 100000) * 0.30
print(f"Total income : ${income:,}")
print(f"Calculated Tax : ${tax:,}")
print(f"Take-Home Income :{income - tax:,}")
```

```python
# 54. Implement a grading system using nested if–else.# 1. Input score
score = 85
# 2. Outer If-Else: First, check if the score is physically valid
if score >= 0 and score <= 100:
    print(f"Score {score} is valid. Processing grade...")
    # Inner Nested If-Else: Evaluate the grade boundaries
    if score >= 90:
        grade = "A"
    else:
        if score >= 80:
            grade = "B"
        else:
            if score >= 70:
                grade = "C"
            else:
                if score >= 60:
                    grade = "D"
                else:
                    grade = "F"
                    
    print(f"Final Grade: {grade}")
else:
    # Triggers if the score is out of the 0-100 range
    print("Error: Invalid score! Marks must be between 0 and 100.")
```

```python
# 55. Validate strong password using multiple conditions.
# 1. Input Password
password = "Secure@Pass123"
# 2. Initialize tracking flags
has_upper = False
has_lower = False
has_digit = False
has_special = False
# Define a string of common special characters
special_characters = "!@#$%^&*()-_=+[{]};:'\",<.>/?`~"
# 3. Check length first using an 'and' logic gate
is_long_enough = len(password) >= 8
# 4. Loop through each character to check types
for char in password:
    if char.isupper():
        has_upper = True
    elif char.islower():
        has_lower = True
    elif char.isdigit():
        has_digit = True
    elif char in special_characters:
        has_special = True
# 5. Final validation using AND
if is_long_enough and has_upper and has_lower and has_digit and has_special:
    print("Password Strength: STRONG")
else:
    print("Password Strength: WEAK")
    print("\nMissing Requirements:")
    if not is_long_enough: print("- Must be at least 8 characters long")
    if not has_upper:      print("- Missing an uppercase letter (A-Z)")
    if not has_lower:      print("- Missing a lowercase letter (a-z)")
    if not has_digit:      print("- Missing a number (0-9)")
    if not has_special:    print("- Missing a special character (e.g., @, #, $)")
```

```python
#56. Calculate delivery charges based on location and order amount.
# 1. Input variables
location = "Domestic"  # Can be "Domestic" or "International"
order_amount = 45.50
# 2. Initialize delivery charge variable
delivery_charge = 0
# 3. Calculate charges using nested if-else and logical operators
if location == "Domestic":
    if order_amount > 50:
        delivery_charge = 0
    else:
        delivery_charge = 5
elif location == "International":
    if order_amount > 100:
        delivery_charge = 15
    else:
        delivery_charge = 25
else:
    print("Error: Unknown shipping location location.")
    delivery_charge = -1  # Error flag
# 4. Print the final bill details if location was valid
if delivery_charge != -1:
    total_bill = order_amount + delivery_charge
    print(f"Location: {location}")
    print(f"Order Amount: ${order_amount:.2f}")
    print(f"Delivery Charge: ${delivery_charge:.2f}")
    print("-" * 30)
    print(f"Total Bill: ${total_bill:.2f}")
```

```python
# 58. Create movie ticket pricing logic based on age & show time.
# 1. Inputs
age = 30
show_hour = 14  # Using 24-hour format (e.g., 14 means 2:00 PM, 19 means 7:00 PM)
# 2. Base standard adult ticket price
ticket_price = 0
# 3. Determine pricing using logic gates (and, or)
is_special_age_group = (age < 12) or (age >= 65)
is_matinee = (show_hour < 17)
if is_special_age_group:
    # Children and seniors get a flat discount rate
    ticket_price = 8
else:
    # Standard adult pricing rules
    if is_matinee:
        ticket_price = 11  # Afternoon discount
    else:
        ticket_price = 15  # Regular peak evening price
# 4. Display the output
print(f"Age of patron: {age}")
print(f"Show Time Hour: {show_hour}:00")
print("-" * 25)
print(f"Final Ticket Price: ${ticket_price}")
```

```python
# 60. Create a menu-driven program using if–elif–else.
# 1. Initialize variables
user_choice = 0
total_bill = 0
# 2. Main program loop
while user_choice != 4:
    # Display the interactive menu
    print("\n========= COFFEE SHOP MENU =========")
    print("1. Espresso        - $3.50")
    print("2. Cappuccino      - $4.50")
    print("3. Chocolate Muffin - $2.50")
    print("4. Checkout & Exit")
    print("====================================")   
    # Take user input
    user_choice = int(input("Enter your choice (1-4): "))   
    # 3. Process choice using if-elif-else
    if user_choice == 1:
        print(">> Added Espresso to your cart.")
        total_bill = total_bill + 3.50
        print(f"Current Total: ${total_bill:.2f}")        
    elif user_choice == 2:
        print(">> Added Cappuccino to your cart.")
        total_bill = total_bill + 4.50
        print(f"Current Total: ${total_bill:.2f}")    
    elif user_choice == 3:
        print(">> Added Chocolate Muffin to your cart.")
        total_bill = total_bill + 2.50
        print(f"Current Total: ${total_bill:.2f}")    
    elif user_choice == 4:
        # Exit condition handling
        print("\n==============================")
        print("        FINAL RECEIPT        ")
        print(f"Thank you! Your total bill is: ${total_bill:.2f}")
        print("Have a great day!")
        print("==============================")     
    else:
        # Fallback for invalid inputs
        print(">> Invalid option! Please select a valid number between 1 and 4.")
```
