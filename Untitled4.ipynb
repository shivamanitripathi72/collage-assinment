```python
# 26. Calculate electricity bill using slab-wise rates.
# Yahan aap apne total units daal sakte hain
units = 250
bill = 0
# Case 1: Agar units 100 ya usse kam hain
if units <= 100:
    bill = units * 3
# Case 2: Agar units 100 se zyada hain lekin 200 ya usse kam hain
else:
    if units <= 200:
        bill = (100 * 3 ) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) +((units - 200) * 8)
# Final bill print karein
print(f"Total units consumed: {units}")
print(f"Aapka Electricity Bill hai: ₹{bill}")
```

```python
# 38. Print discount percentage based on purchase amount.
# Step 1: Define the purchase amount
purchase_amount = 250  # Change this value to test different discounts
# Step 2: Procedural logic to determine the discount percentage
if purchase_amount < 50:
    discount_percent = 0
elif purchase_amount <= 100:
    discount_percent = 5
elif purchase_amount <= 500:
    discount_percent = 10
else:
    discount_percent = 20
# Step 3: Calculate the savings and final price
savings = purchase_amount * (discount_percent / 100)
final_price = purchase_amount - savings
# Step 4: Output the results
print("Purchase Amount: $", purchase_amount, sep="")
print("Discount Applied: ", discount_percent, "%", sep="")
print("Total Savings:   $", savings, sep="")
print("Final Bill:      $", final_price, sep="")
```
