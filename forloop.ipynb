```python
#21. Write a Python program to find the longest increasing subsequence in a list.
# Longest Increasing Subsequence without using a function
seq = [10, 22, 9, 33, 21, 50, 41, 60]
n = len(seq)
dp = [1] * n          # हर index पर LIS की length
parent = [-1] * n     # subsequence reconstruct करने के लिए
for i in range(1, n):
    for j in range(0, i):
        if seq[j] < seq[i] and dp[j] + 1 > dp[i]:
            dp[i] = dp[j] + 1
            parent[i] = j
# सबसे लंबी subsequence का index
max_len = max(dp)
max_index = dp.index(max_len)
# subsequence reconstruct करना
lis = []
while max_index != -1:
    lis.append(seq[max_index])
    max_index = parent[max_index]
lis.reverse()
print("Original list:", seq)
print("Longest Increasing Subsequence:", lis)
```

```python
# 22. Write a program to group elements based on frequency.
# Group elements based on frequency

numbers = [1, 2, 2, 3, 4, 4, 4, 5, 1, 3, 3]

# Step 1: Count frequency
freq = {}
for num in numbers:
    freq[num] = freq.get(num, 0) + 1

# Step 2: Group by frequency
grouped = {}
for num, count in freq.items():
    if count not in grouped:
        grouped[count] = []
    grouped[count].append(num)

# Print results
print("Original list:", numbers)
print("Frequency count:", freq)
print("Grouped by frequency:", grouped)
```

```python
"""Tricky
11. Invert a dictionary (swap keys and values).
12. Group elements by frequency using a dictionary.
13. Find duplicate values in a dictionary.
14. Create a nested dictionary for student records.
15. Flatten a nested dictionary."""

# Invert dictionary
my_dict = {"a": 1, "b": 2, "c": 3}

inverted = {v: k for k, v in my_dict.items()}

print("Original dictionary:", my_dict)
print("Inverted dictionary:", inverted)
print("---"*28)

# Group by frequency
my_list = [1, 2, 2, 3, 4, 4, 4, 5]

freq = {}
for item in my_list:
    freq[item] = freq.get(item, 0) + 1

grouped = {}
for key, value in freq.items():
    grouped.setdefault(value, []).append(key)

print("Original list:", my_list)
print("Grouped by frequency:", grouped)
print("---"*28)

# Duplicate values
my_dict = {"a": 1, "b": 2, "c": 1, "d": 3, "e": 2}

values = list(my_dict.values())
duplicates = set([v for v in values if values.count(v) > 1])

print("Dictionary:", my_dict)
print("Duplicate values:", duplicates)
print("---"*28)

# Nested dictionary for students
students = {
    "101": {"name": "Shiva", "age": 21, "course": "Python"},
    "102": {"name": "Aman", "age": 22, "course": "Java"},
    "103": {"name": "Riya", "age": 20, "course": "C++"}
}

print("Student Records:")
for roll, info in students.items():
    print(f"Roll No: {roll}, Name: {info['name']}, Age: {info['age']}, Course: {info['course']}")
print()
print("---"*28)

# Flatten nested dictionary
nested_dict = {
    "student": {
        "name": "Shiva",
        "details": {
            "age": 21,
            "course": "Python"
        }
    }
}

flat_dict = {}

def flatten(d, parent_key=""):
    for k, v in d.items():
        new_key = parent_key + "." + k if parent_key else k
        if isinstance(v, dict):
            flatten(v, new_key)
        else:
            flat_dict[new_key] = v

flatten(nested_dict)

print("Nested dictionary:", nested_dict)
print("Flattened dictionary:", flat_dict)
print("---"*28)
```
